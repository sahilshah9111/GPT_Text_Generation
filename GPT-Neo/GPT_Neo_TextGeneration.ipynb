```python
from transformers import GPTNeoForCausalLM, GPT2Tokenizer
import gradio as gr
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("EleutherAI/gpt-neo-1.3B")
model = GPTNeoForCausalLM.from_pretrained("EleutherAI/gpt-neo-1.3B")
```

```python
def generate_text(inp):
  input_ids = tokenizer(inp, return_tensors = 'pt').input_ids
  beam_output = model.generate(input_ids, min_length = 200, max_length = 300, do_sample = True, temperature=0.9)
  gen_text = tokenizer.batch_decode(beam_output)[0]
  return gen_text
```

```python
output_text = gr.Textbox()
gr.Interface(generate_text,"textbox", output_text, title = "GPT-Neo Text generation",
             description = "OpenAI's GPT-Neo Large is a powerful model. It is unsupervised \
             language model that can generate coherent text. Just input any text and see \
             for yourself! It's takes a couple of minutes, so please be patient!").launch()
```
